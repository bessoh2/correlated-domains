### Make a fishnet from the data

```python
# Code adapted from https://spatial-dev.guru/2022/05/22/create-fishnet-grid-using-geopandas-and-shapely/

from shapely import geometry

# Get minX, minY, maxX, maxY
minX, minY, maxX, maxY = aso_500m.rio.bounds()

# Create a fishnet
x, y = (minX, minY)
geom_array = []

# Polygon Size
square_size = 500
while y <= maxY:
    while x <= maxX:
        geom = geometry.Polygon([(x,y), (x, y+square_size), (x+square_size, y+square_size), (x+square_size, y), (x, y)])
        geom_array.append(geom)
        x += square_size
    x = minX
    y += square_size

fishnet = gpd.GeoDataFrame(geom_array, columns=['geometry']).set_crs('EPSG:32611')
fishnet.to_file('fishnet_grid.shp')
```

```python
# Plot the fishnet on top of the coarsened ASO data:
fig, ax = plt.subplots()

aso_500m.plot(ax=ax)
fishnet.plot(ax=ax, edgecolor='k', facecolor='none')
```

```python
from rasterstats import zonal_stats

fishnet['aso_median'] = np.nan

for i in range(len(fishnet)):
    rasterstats = zonal_stats(fishnet.loc[i,'geometry'], aso_fn, nodata=(-9999.0), 
                          stats="median")
    fishnet.loc[i,'aso_median'] = rasterstats[0]['median']

fishnet['aso_median'] = fishnet['aso_median'].astype('float')
```

### Now run the same method for all 8 pairs:

```python
for aso_date, is2_date, aso_fn in zip(aso_dates,atl06_dates, aso_fn_winters):
    fishnet[f'aso_median_{aso_date}'] = np.nan
    fishnet[f'is2_median_{is2_date}'] = gpd.sjoin(atl06sr_gdf[atl06sr_gdf['acqdate']==is2_date], fishnet, how='left').groupby('index_right')['diff_plus_bicubic'].median()
    index_list = list(gpd.sjoin( atl06sr_gdf[atl06sr_gdf['acqdate'] == is2_date], fishnet, how='left').index_right.unique())
    
    for i in index_list:
        rasterstats = zonal_stats(fishnet.loc[i,'geometry'], aso_fn, nodata=(-9999.0), 
                              stats="median")
        fishnet.loc[i,f'aso_median_{aso_date}'] = rasterstats[0]['median']

    fishnet[f'aso_median_{aso_date}'] = fishnet[f'aso_median_{aso_date}'].astype('float')
    fishnet[f'difference_{aso_date}_{is2_date}'] = fishnet[f'aso_median_{aso_date}'] - fishnet[f'is2_median_{is2_date}']
```

```python
fig, ax = plt.subplots(1,4, figsize=(15,5))
fishnet.plot(column='aso_median_2019-04-17',legend=True, ax=ax[0])
fishnet.plot(column='is2_median_2019-04-10',legend=True, ax=ax[1])
fishnet.plot(column='difference_2019-04-17_2019-04-10',legend=True, ax=ax[2])
fishnet['difference_2019-04-17_2019-04-10'].hist(bins = int(np.sqrt(90)), ax=ax[3])
```
