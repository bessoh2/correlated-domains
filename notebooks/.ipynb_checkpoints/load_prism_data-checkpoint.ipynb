### Load in useful functions for working with .bil and .hdf files, from:  
PyMorton: https://pymorton.wordpress.com/2016/03/04/adding-state-outlines-to-prism-figures-without-using-basemap/

```python
def read_prism_hdr(hdr_path):
    """Read an ESRI BIL HDR file"""
    with open(hdr_path, 'r') as input_f:
        header_list = input_f.readlines()
    return dict(item.strip().split() for item in header_list)
 
def read_prism_bil(prism_path):
    """Read an array from ESRI BIL raster file"""
    header_dict = read_prism_hdr(prism_path.replace('.bil', '.hdr'))
    prism_nodata = float(header_dict['NODATA'])
    prism_rows = int(header_dict['NROWS'])
    prism_cols = int(header_dict['NCOLS'])
    
    # Assume the cells are square and have the same cellsize in each direction
    prism_cs = float(header_dict['XDIM'])
    # prism_cs = float(header_dict['YDIM'])
    
    # Build extent as xmin, xmax, ymin, ymax
    # ULXMAP and ULYMAP are upper left cell center
    prism_extent = [
        float(header_dict['ULXMAP']) - 0.5 * prism_cs,
        float(header_dict['ULXMAP']) - 0.5 * prism_cs + prism_cols * prism_cs,
        float(header_dict['ULYMAP']) + 0.5 * prism_cs - prism_rows * prism_cs,
        float(header_dict['ULYMAP']) + 0.5 * prism_cs
    ]

    # Read in the array
    prism_array = np.fromfile(prism_path, dtype=np.float32)
    prism_array = prism_array.reshape(prism_rows, prism_cols)
    prism_array[prism_array == prism_nodata] = np.nan
    return prism_array
```

### Read all the files and populate a list with these arrays:

```python
precip_data = []

for file in files:
    precip_data.append(read_prism_bil(file))
```

### Make a dictionary with time, x, y, and precip data - then convert this to an xarray dataarray:

```python
d = {
    "coords": {
        "time": {"dims": "time", "data": xr.cftime_range(start="1989-01", end="2021-12", freq="MS", calendar="noleap")},
        "y": {"dims":"y", "data":ys},
        "x": {"dims": "x", "data": xs}
    },
    "attrs": {"units": "millimeters)"},
    #"dims": "date",
    "data": precip_data
}

```

```python
da = xr.DataArray.from_dict(d)
```

```python
da = da.rio.write_crs('epsg:4326')
```

### Plot this with the western states and UCRB to check that the coordinates look right:

### Clip to the UCRB, our study area:

```python
ucrb_precip = da.rio.clip(ucrb.geometry)
```

```python
ucrb_precip.sel(time='1990-03').plot()
```

### Calculate total winter precipitation in each water year:

```python
cs_precip = xr.full_like(ucrb_precip, np.nan)
```

```python
cs_precip = cs_precip.resample(time='YS-APR').mean()
```

```python
cs_precip = cs_precip.sel(time=slice('1989','2021'))
```

```python
for year in np.arange(1990,2022):
    cs_precip.loc[dict(time=f'{year}')] = (ucrb_precip.where((ucrb_precip.sel(time=(f'{year-1}')).time.dt.month >= 11)).sum(dim='time') + ucrb_precip.where((ucrb_precip.sel(time=f'{year}').time.dt.month <= 3)).sum(dim='time'))
```

```python
cs_precip.sel(time='1992').plot()
```

```python
cs_precip = cs_precip / 1000
```
