### Load and Explore Priors

```python
priors = xr.open_dataset('../data/reanalysis/Priors/Prior_April1st_SWE_WYs_1990_2021.nc')
```

```python
priors
```

```python
priors = priors.rename({'lat':'y','lon':'x','latitude':'y', 'longitude':'x'})
```

```python
priors = priors.squeeze()
```

```python
priors = priors.rio.set_spatial_dims('y','x',inplace=True)
```

```python
priors = priors.SWE.assign_coords(x=priors['x'], y=priors['y'])
priors = priors.T
```
